```julia
using GraphTools
```

```julia
g = (@graph A -> B -> C -> D) + (@graph C 50-> A) + (@graph P -> Q)
g.attributes["splines"] = "curved"
g.attributes["layout"] = "fdp"
g.attributes["K"] = 0.5
draw(g)
```

```julia
b = vertexof(g, "B")
b["style"] = "filled"
b["fontcolor"] = "black"
b["fillcolor"] = "yellow"
b["color"] = "red"
draw(g, b)
```

```julia
c = vertexof(g, "C")
c["style"] = "filled"
c["shape"] = "box"
c["fillcolor"] = "orange"
draw(g, c)
```

```julia
e = edgeof(g, "B", "C")
e["len"] = 1.5
e["color"] = "cyan"
draw(g, e)
```

```julia
draw(g)
```

```julia
p = vertexof(g, "P")
p["style"] = "filled"
p["fontcolor"] = "black"
p["fillcolor"] = "yellow"
p["color"] = "red"
draw(g, p)
```

```julia
for_vertices(g) do v
    if !haskey(v.attributes, "style")
        v["style"] = "filled"
        v["fillcolor"] = "snow"
    end
end

for_edges(g) do e
    if !haskey(e.attributes, "color")
        e["color"] = "brown"
    end
end

draw(g)
```
