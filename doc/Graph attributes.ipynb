```julia
using GraphTools
import GraphViz: Graph
import Graphs: ExVertex, ExEdge

draw(g::GraphStore) = Graph(to_dot(g))
draw(g::GraphStore, ex::Union{ExVertex,ExEdge}) = Graph(to_dot(g, ex))
```

```julia
g = (@graph A 50-> B -> C -> D) + (@graph C -> G -> H -> A) + (@graph I -> J -> K)
g.attributes["splines"] = "curved"
g.attributes["layout"] = "fdp"
g.attributes["K"] = 0.5
draw(g)
```

```julia
b = vertexof(g, "B")
b["style"] = "filled"
b["fontcolor"] = "black"
b["fillcolor"] = "yellow"
b["color"] = "red"
draw(g, b)
```

```julia
c = vertexof(g, "C")
c["style"] = "filled"
c["shape"] = "box"
c["fillcolor"] = "orange"
draw(g, c)
```

```julia
e = edgeof(g, "B", "C")
e["len"] = 1.5
e["color"] = "cyan"
draw(g, e)
```
