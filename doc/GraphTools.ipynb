```julia
using GraphTools
import GraphViz: Graph
import Graphs: ExVertex, ExEdge

draw(g::GraphStore) = Graph(to_dot(g))
draw(g::GraphStore, ex::Union{ExVertex,ExEdge}) = Graph(to_dot(g, ex))
```

```julia
draw( @graph A )
```

```julia
draw( @graph A - B )
```

```julia
draw( @graph A -> B )
```

```julia
draw( @graph A 50-> B )
```

```julia
draw( @graph A 50-> B -> C )
```

```julia
draw( @graph A -> B -> C -> A )
```

```julia
draw( (@graph A -> B) + (@graph A -> C) + (@graph D) )
```
